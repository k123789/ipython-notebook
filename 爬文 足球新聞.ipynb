```python
import requests
from bs4 import BeautifulSoup
import json
```

```python
for round in range(1,29):
    url= "https://www.foootball.cc/blog?category=foootballnews&page={}".format(round)
    res=requests.get(url)
    soup=BeautifulSoup(res.text,'html.parser')
    data=soup.find_all('h3')
    for i in data:
        print(i.text)
```
