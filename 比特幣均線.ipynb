```python
import requests
from bs4 import BeautifulSoup
res = requests.get('https://www.coingecko.com/zh-tw/%E5%8C%AF%E7%8E%87%E5%9C%96/%E6%AF%94%E7%89%B9%E5%B9%A3/twd')
soup= BeautifulSoup(res.text,'html.parser')
```

```python
data_prices = soup.select('#coin_portfolio_price_chart_btc')[0].prettify('utf-8').decode('utf-8')
```

```python
import re
m = re.search('<div class="coin_portfolio_price_chart" data-prices="(.*?)"',data_prices)
```

```python
import json
jd=json.loads(m.group(1))
jd
```

```python
import pandas
df = pandas.DataFrame(jd)
```

```python
df.columns = ['datetime','twd']
```

```python
df['datetime']=pandas.to_datetime(df['datetime'],unit='ms')
```

```python
df.head(100)
```

```python
df.index=df['datetime']
```

```python

df.head()
```

```python
%pylab inline
df['twd'].plot(kind='line',figsize=[10,5])
```

```python
df['ma7']=df['twd'].rolling(window=7).mean()
```

```python
df.head(20)
```

```python
df[['twd','ma7']].plot(kind='line',figsize=[10,5])
```

```python
df2=df[df['datetime']>='07:30']
```
